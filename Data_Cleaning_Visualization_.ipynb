```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("StudentsPerformance.csv")

print(df.info())
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
print("Duplicate Rows:")
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates()

print("Duplicates Removed")
print(df.shape)
```

```python
print("Average Math Score:")
print(df["math score"].mean())

print("Average Reading Score:")
print(df["reading score"].mean())

print("Average Writing Score:")
print(df["writing score"].mean())
```

```python
df["gender"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")

plt.savefig("gender_chart.png")

plt.show()

```

```python
df["math score"].plot(kind="hist", bins=10)

plt.title("Math Score Distribution")

plt.savefig("score_histogram.png")

plt.show()
```
